### **Cargar Librerías** 

```python
import pandas as pd
import seaborn as sns 
from matplotlib import pyplot as plt
```

### **Caso 1** 

```python
df1 = pd.read_csv("Tareas Diplomado/videoconferencia.csv").pivot(index="Día",columns="Semana",values="Horas")
df1=df1.reindex(['Lunes','Martes','Miércoles','Jueves','Viernes','Sábado','Domingo'])
df1
```

```python
plt.rcParams['figure.figsize']=(12.0,8.0)

hm1 = sns.heatmap(df1)
```

```python
#Divergente con anotaciones

hm1 = sns.heatmap(df1, annot=True,fmt='d',annot_kws={'size':8},cmap='coolwarm',linewidths=0.8,square=True)
```

```python
#Secuencial con anotaciones

hm1 = sns.heatmap(df1, annot=True,fmt='d',annot_kws={'size':8},cmap='Blues',linewidths=0.8,square=True)
```

Dado que no existen valores negativos para la magnitud de los valores correspondientes a los colores, ni tampoco un centro definido, es preferible utilizar una paleta de colores de tipo secuencial. Con ello, es más fácil observar observar las diferencias de magnitud entre los valores.

```python
data1=[136,87,119]
labels=["Primera Clase","Segunda Clase","Tercera Clase"]
colors=["Yellow","Purple","Pink"]
fig=plt.figure(figsize=(8,8))

data2=[80,97,372]
labels=["Primera Clase","Segunda Clase","Tercera Clase"]
colors=["Yellow","Purple","Pink"]
fig=plt.figure(figsize=(8,8))

plt.subplot(1,2,1)
plt.pie(data1,labels=labels, autopct='%1.1f%%',colors=colors)
plt.title("“Pasajeros del Titanic por Supervivencia (0/1) y por Clase")
plt.xlabel('Supervivientes')

plt.subplot(1,2,2)
plt.pie(data2,labels=labels, autopct='%1.1f%%',colors=colors)
plt.xlabel('No Supervivientes')

plt.show()
```

pasajero más común?" se puede evidenciar mediante el gráfico que los pasajeros de tercera clase son aquellos con mayor tasa de no supervivencia, y de la pregunta "c) ¿cree que este gráfico es más efectivo que un gráfico de barras para comparar categorías?" se puede arguir que el gráfico de torta permite percibir con mayor facilidad las proporciones de las magnitudes de las variables medidas. Por ello, en este caso, sería más provechoso quedarse con gráficos de torta.
