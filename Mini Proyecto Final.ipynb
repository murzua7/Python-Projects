```python
# V07: Visualización que permita apreciar distribución y potencialmente correlación entre pares de atributos de edad, peso y altura; 
# con categorización por tipos de síntomas presentados. -> gráfico de burbujas
# acción: analizar-derivar
# objetivo: atributo-distribución/correlación
# Dataset: V07.csv 

df7 = pd.read_csv(r"C:\Users\Matias U\Desktop\V07.csv")
df7
```

```python
sint = df7.Sintomas.astype("category").cat.codes
```

```python
sns.set_style("whitegrid")

ax7 = sns.scatterplot(x="Peso",y="Altura",hue=sint,size="Edad",sizes=(10,400),hue_norm=(0,1.5),alpha=0.5,cmap="YlOrRd", legend="brief",data=df7)
ax7.set_title("Peso vs Altura \n (tamaño de burbuja: edad) \n (color: síntoma base o extendido)")
```

**Justificación**

La acción correspondiente a esta última visualización es de analiza y derivar, mientras que el objetivo es de ver la la correlación entre atributos presentes en el dataset. Se considera esta acción ya que se desea generar nueva información a partir de la gráfica al mostrar las cuatro variables al unísono, derivando una mejor descripción de los perfiles y la correlación de las variables. El objetivo es ver la corrrelación entre las variables para armar perfiles de personas más precisos. Para cumplir estas tareas se utilizó un gráfico de burbujas, ya que permite la visualización de 4 variables sin mayores problemas.
